# Exploring New Year's Resolution Tweets

This project involves building an unsupervised learning model to identify the primary topics that make up New Year's resolutions and the general sentiment surrounding these resolutions. The goal is to identify what areas of the wellness field are best promoted at this time of year so that we can help individuals better improve their health and well-being.

### Topic Modeling

The initial stages of topic modeling have shown there is a wide range of tweets that include mention of "new years resolution" or "2022 resolution." As such, I have used NMF to identify potential topics and the top words associated with those topics. I have then used some of these words within CorEx to improve on the topic modeling.
